```python
products = ["t-shirt", "mug", "hat", "book", "keychain"]

def initialise_inventory(products):
    inventory = {product: int(input(f"Enter the quantity of {product}s available: ")) for product in products}
    return inventory

    customer_orders = set()

    more_order = "yes"

def get_customer_orders():
    number_orders = int(input("Enter the number of customer orders: "))
    customer_orders = {input(f"Enter product name for order {i+1}: ") for i in range(number_orders)}
    return customer_orders

def calculate_total_price(customer_orders):
    prices = {product: float(input(f"Enter the price for {product}: ")) for product in customer_orders}
    total_price = sum(prices[product] for product in customer_orders)
    return total_price

def update_inventory(customer_orders, inventory):
    updated_inventory = {
        item: inventory[item] - 1 if item in customer_orders else inventory[item]
        for item in inventory
    }
    
    filtered_inventory = {
        item: qty for item, qty in updated_inventory.items() if qty > 0
    }
    return filtered_inventory

def calculate_order_statistics(customer_orders, products):
    total_ordered = len(customer_orders)
    percent_ordered = (total_ordered / len(products)) * 100
    return total_ordered, percent_ordered


def print_order_statistics(order_statistics):
    print("Order Statistics:", order_statistics)
    print("Total Products Ordered:", order_statistics[0])
    print("Percentage of Products Ordered:", f"{order_statistics[1]:.2f}%")

def print_updated_inventory(inventory):
    [print(f"{product}: {quantity}") for product, quantity in inventory.items()]


inventory = initialise_inventory(products)

customer_orders = get_customer_orders()

inventory = update_inventory(customer_orders, inventory)

order_stats = calculate_order_statistics(customer_orders, products)
print("\nOrder Statistics:")
print_order_statistics(order_stats)

print("\nUpdated Inventory:")
print_updated_inventory(inventory)

total_price = calculate_total_price(customer_orders)
print(f"Total Price: {total_price:.2f}")
```
